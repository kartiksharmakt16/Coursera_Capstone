## Introduction: Business Problem
Me and my friend are planning to buy a suitable place in Mumbai to open a coffee shop. We're interesting in a location where there are least No. of coffee shops already present, so that we can maximize the profits.<br>
Mumbai is a costly city and since I'm not much aware about the city, I decided to do a quick study of different areas in Mumbai to see if I can make a descent investment.
<br>

In this project,I will try to find an optimal location to set-up a coffee hangout place.
<br>
Specifically, this report will be targeted to stakeholders interested in opening an Coffee Hangout in Mumbai, India.
<br><br>
Since there are lots of coffee hangouts in Mumbai, I will try to detect locations that are not already crowded with coffee joints.<br>
I'll be using Data Science techniques to generate different visualizations, in forms of graphs and maps, depicting the promissing neighborhoods to open a coffee join in Mumbai.
